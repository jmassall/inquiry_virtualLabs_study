```python
import pandas as pd
import numpy as np
import scipy.stats as stats
# import seaborn as sns
# import matplotlib.pyplot as plt
# import matplotlib
# %matplotlib inline
# matplotlib.rcParams['figure.figsize'] = (10, 3)
```

```python
#load IRR spreadsheets
csvfile1 = 'Interrater - Interrater CB.csv' 
df1 = pd.read_csv(csvfile1, sep=',')

for col in df1.columns:
    if df1[col].dtypes == 'float64':
        df1[col] = df1[col].fillna(-88)
        df1[col] = df1[col].astype(int)
        df1[col] = df1[col].astype(str)
        df1[col] = df1[col].replace('-88', np.nan)
        
df1.fillna(value='NA', inplace=True)
df1= df1.astype(str)

print df1.head()
print df1.dtypes
print df1.shape
```

```python
#load IRR spreadsheets
csvfile2 = 'Interrater - Interrater JM.csv' 
df2 = pd.read_csv(csvfile2, sep=',')

for col in df2.columns:
    if df2[col].dtypes == 'float64':
        df2[col] = df2[col].fillna(-88)
        df2[col] = df2[col].astype(int)
        df2[col] = df2[col].astype(str)
        df2[col] = df2[col].replace('-88', np.nan)

df2.fillna(value='NA', inplace=True)
df2= df2.astype(str)

# print df2.head()
print df2.dtypes
# print df2.shape
```

```python
student_info_headers = df1.columns[0:4]
code_headers = df1.columns[5:]
print student_info_headers, code_headers
```

```python
valid_codes = ['1','0','-1','+1','-0','+0','NA']
for code in code_headers:
    print code 
    print df1[code].unique()
    print df2[code].unique()
```

```python
new_cols1 = [col+'_CB' for col in df1.columns]
df1.columns = new_cols1
new_cols2 = [col+'_JM' for col in df2.columns]
df2.columns = new_cols2
```

```python
df = pd.concat([df1,df2],axis=1)
print df.columns
print df.shape
```

```python
df.head()
```

```python
IRR_percent_dict = {}
IRR_kappa_dict = {}

for code in code_headers:
#     code = df1.columns[5]
    tab = pd.crosstab(df[code+'_CB'], df[code+'_JM'])
    tab_margins = pd.crosstab(df[code+'_CB'], df[code+'_JM'], margins = True)
    print tab_margins
    common_entries = list(set(tab.columns) & set(tab.index))
    print common_entries
    
    agree = 0
    expected = 0
    N = float(tab_margins['All']['All'])
    for entry in common_entries:
        agree += tab[entry][entry]
        expected += tab_margins[entry]['All']*tab_margins['All'][entry] / N
    #     print agree, expected

    IRR_percent_dict[code] = agree/N
    IRR_kappa_dict[code] = (agree - expected)/(N - expected)

    print 'Percent agreement  =', IRR_percent_dict[code]
    print 'Cohen\'s kappa =', IRR_kappa_dict[code], '\n'
```
